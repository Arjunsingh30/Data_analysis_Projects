```python
# Importing required libraries
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Creating an empty DataFrame to store all the data
all_data = pd.DataFrame()
# Defining the path to the directory containing the sales data files
path = 'C:\\Users\\ARJUN\\OneDrive\\Desktop\\Real_world Projects\\sales_data_analysis(Ecommerce_case_study)\\Sales_Data'
for file in files:
    current_df = pd.read_csv(path+'/'+file)     # Reading the current file as a DataFrame
    all_data = pd.concat([all_data,current_df])
```

```python
# Defining a function called 'month' that takes a date string as input and returns the month part of the date.
# The function uses the 'split()' method to split the date string at '/' and extracts the first part (month) using index 0.
def month(date):
    return date.split('/')[0]
```

```python
# The 'apply()' method is used to apply the 'month' function to each element in the 'Order Date' column and obtain the month part.
all_data['month'] = all_data['Order Date'].apply(month)
```

```python
# Converting the 'month' column to integer data type.
all_data['month'] = all_data['month'].astype('int')
```

```python
# Converting the 'Order Date' column to a pandas datetime data type and then extracting the hour from the datetime.
all_data['Hour'] = pd.to_datetime(all_data['Order Date']).dt.hour
all_data['Hour'].head()
```

```python
# Initializing empty lists 'keys' and 'hour' to store the hour values and the count of product purchases for each hour.
keys = []
hour = []

# Grouping the 'all_data' DataFrame by 'Hour' (hour of the day) and iterating over each group.
for key, hour_df in all_data.groupby('Hour'):
    keys.append(key)
    hour.append(len(hour_df))
```

```python
# Creating a line plot to visualize the number of product purchases at different hours of the day.
plt.grid()
plt.plot(keys, hour)
plt.xlabel('Hours')
plt.ylabel('Product Purchases')
plt.xticks(keys)
```

```python
# Creating a new column 'grouped' in the 'df' DataFrame that contains a comma-separated string of products for each 'Order ID'.
df['grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
df.head()
```
